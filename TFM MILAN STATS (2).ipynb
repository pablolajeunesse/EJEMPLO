#             <span style="color: red;">DATA SCIENCE PROJECT: RECONSTRUCTING AC MILAN</span>

Not too long ago, AC Milan was considered the best team in Italy and one of the best teams in the world. With a total of 52 official titles including 18 Serie A, 2 Serie B, 5 Italian Cups, 7 Italian Supercups, 7 UEFA Champions Leagues (making them the second team with most Champions Leagues in history, only falling behind Real Madrid with 13), 3 intercontinental Cups, 1 FIFA Club World Cup, 2 UEFA Cup Winners Cups amongst other important trophies, this historical club had always had the best football superstars when it came to players and coaches, and had must definitely always been a team to fear when played against. 

However, this winning tendency started to change approximately 10-11 years ago after their last italian Serie A title in the 2010-11 season (on that year, they also lifted the italian Supercup). Since that season, AC Milan has only won one italian Supercup in 2016 and has been to the italian Cup final twice (2015-16 and 2017-18), losing both times.

After their last winning season in 2010-11 the team never changed nor rightfully reconstructed its roster, while other teams like Juventus, NÃ¡poli or even Roma kept building their way up, signing and bringing great players and thus improving their squads. Then, they slowly started losing their best players (like Zlatan Ibrahimovic in attack and Thiago Silva in defence), and the economic crisis hit them hard, leaving them with financial problems and less money to improve their team during the transfer windows.

Today, the reality is that AC Milan is no more than a struggling club which last qualified to UEFA Champions League in the 2012-2013 season (so 7 years ago, when they finished 3rd in Italian Serie A, which was the last spot required in order to qualify for the prestigious European tournament for the next season, until the rules changed in 2017-2018 allowing the 4th in the table to qualify as well), and most importantly a club which has completely lost the respect and fear effect from its rivals.

The objective of this project will be to use data analysis to try and find out what AC Milan needs (footballingly and statistically speaking) to improve in order to make it back to UEFA Champions League spots in italian Serie A (only the top 3 teams of each league qualify to Champions League each season). We will analyze, compare, and visualize AC Milan's football statistics from the last time they qualified to UEFA Champions League (2012-13 season) until their last completed season (2018-2019). 
We will also compare Milan's stats to those of teams which did qualify to UEFA Champions League, using them as a reference and Key Performance Indicators Milan will have to imitate and reach in order to get back to the prestigious european competition.

## ANALYZING/VIZUALISING/COMPARING DATA:

```python
#Representation of Milans total points vs those of the 3rd in the league for each season. 
#We will call these teams "UCL team" as 3rd in the table is the limit/last spot to qualify for Champions League:

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure


plt.figure(figsize=(16,5))
plt.plot(Data_ACMilan['Season'], Data_ACMilan['Total points'], color='r', marker='o', label='Milans points')
plt.plot(Data_CLStats['Season'], Data_CLStats['Total points'], color='b', marker='o', label='UCL team points')

plt.title('Milan vs UCL team TOTAL POINTS per season')
plt.xlabel('Seasons')
plt.ylabel('Points')
plt.legend()

for x,y in zip(Data_ACMilan['Season'], Data_ACMilan['Total points']):
    label = "{:} pts".format(y)
    plt.annotate(label, (x,y), textcoords='offset points', xytext=(0, 10), ha='center')
    
for x,y in zip(Data_CLStats['Season'], Data_CLStats['Total points']):
    label = "{:} pts".format(y)
    plt.annotate(label, (x,y), textcoords='offset points', xytext=(0, 10), ha='center')    
```

The plot above is a representation of Milan vs UCL teams total points per season.

The plot above is a representation of Milan vs UCL teams points per game each season.

These first two graphs help us visualize how big the gap between Milan and the 3rd in Serie A league's table was for most of the last 7 seasons. With the following plots, we are going more in depth by showing and comparing Milan vs UCL team stats per season:

```python
#Compare Milans amount of Wins, Losses, Draws each season to those of UCL Teams:

X = Data_ACMilan[['Season', 'Wins', 'Draws', 'Losses']] 
Y = X.set_index('Season')
Z = Y.groupby('Season').sum()

U = Data_CLStats[['Season', 'Wins', 'Draws', 'Losses']]
C = U.set_index('Season')
L = C.groupby('Season').sum()

XYZ = Z.plot.bar(title='Milans Record per season', stacked=True, figsize=(12,8))
UCL = L.plot.bar(title='UCL team Record per season', stacked=True, figsize=(12,8))

for p in XYZ.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    XYZ.text(x+width/2, 
            y+height/2, 
            '{:}'.format(height), 
            horizontalalignment='center', 
            verticalalignment='center')
    
for p in UCL.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    UCL.text(x+width/2, 
            y+height/2, 
            '{:}'.format(height), 
            horizontalalignment='center', 
            verticalalignment='center')    
```

The plots above compare Milans amount of Wins, Losses, Draws each season to those of UCL Teams.

Thanks to these plots we can appreciate Milan's and UCL Team's results each season. 

As we can see, Milan hasn't reached 20 wins in a single season since they finished 3rd in 2012-2013, which is when they last qualified for Champions League. On top of that, they lost 10 or more games 5 out of the last 7 seasons. 

On the other side, UCL Teams have reached 20 wins or more every single season and have lost 10 or more games only once (the 2014-2015 season).

It is obvious to say Milan should try to win every single game, but a good and realistic objective they should have when starting the season would be aiming at least the 20 wins mark. It is also important to know that losing 10 or more games in a season makes it really difficult for a team to make it to UCL Champions League spots. 

In a competition where every single point counts, it is important for a team like AC Milan to know these type of statistics in order to achieve its goals/objectives as a club. 
Once Milan reaches 20 wins & less than 10 losses in a season they will have a shot at Champions League spots.

They came really close the 2018-2019 season, finishing in 5th position with 19 wins and 8 losses. Unfortunately for the club, the 2 teams ahead of them ended the season with one more win, leaving them out of european spots.

```python
#We plot the Overall_Stats DataFrame in a polar plot to compare AC Milan vs UCL Teams

import numpy as np
import pylab as pl

class Radar(object):
    def __init__(self, fig, titles, labels, rotation=0, rect=None):
        if rect is None:
            rect = [0.05, 0.05, 0.95, 0.95]

        self.n = len(titles)
        self.angles = np.arange(0, 360, 360.0/self.n)
        self.axes = [fig.add_axes(rect, projection="polar", label="axes%d" % i) 
                         for i in range(self.n)]

        self.ax = self.axes[0]
        self.ax.set_thetagrids(self.angles, labels=titles, fontsize=14)

        for ax in self.axes[1:]:
            ax.patch.set_visible(False)
            ax.grid("off")
            ax.xaxis.set_visible(False)

        for  ax, angle, label in zip(self.axes, self.angles, labels):
            ax.set_rgrids(range(1, 7), angle=angle, labels=label)
            ax.spines["polar"].set_visible(True)
            ax.set_ylim(0, 6)
            ax.set_theta_offset(np.deg2rad(rotation))

    def plot(self, values, *args, **kw):
        angle = np.deg2rad(np.r_[self.angles, self.angles[0]])
        values = np.r_[values, values[0]]
        self.ax.plot(angle, values, *args, **kw)
    
    def fill(self, values, *args, **kw):
        angle = np.deg2rad(np.r_[self.angles, self.angles[0]])
        values = np.r_[values, values[0]]
        self.ax.fill(angle, values, *args, **kw)    



fig = pl.figure(figsize=(8, 8))

titles = ['Points per game','Win %','Goals scored',
          'Goals conceded', 'Failed to score %', 'Shots',
          'Conversion rate of shots on target (%)', 'Conversion rate of shots (%)',
          'Shots on target', 'Average possession (%)']

labels = [[0.5, 1, 1.5, 2, 2.5], 
          [10, 20, 30, 40, 50],
          [30, 40, 50, 60, 70],
          [10, 20, 30, 40, 50],
          [10, 20, 30, 40, 50],
          [100, 200, 300, 400, 500],
          [15, 20, 25, 30, 35],
          [5, 10, 15, 20, 25],
          [210, 220, 230, 240, 250],
          [40, 50, 60, 70, 80]]

radar = Radar(fig, titles, labels)
radar.plot([3.15, 4.317, 3.5, 4.417, 2.2, 4.8967, 3.145, 2.24, 1.367, 2.3], "-", lw=2, color="r", alpha=0.8, label="AC Milan from 2013 to 2019")
radar.fill([3.15, 4.317, 3.5, 4.417, 2.2, 4.8967, 3.145, 2.24, 1.367, 2.3], color="r", alpha=0.1)

radar.plot([4.05, 5.983, 5.717, 3.85, 1.617, 5.4183, 4.41, 2.85, 4.083, 2.567],"-", lw=2, color="b", alpha=0.8, label="UCL teams from 2013 to 2019")
radar.fill([4.05, 5.983, 5.717, 3.85, 1.617, 5.4183, 4.41, 2.85, 4.083, 2.567], color="b", alpha=0.1)


radar.ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.10),
      fancybox=True, shadow=True, ncol=4, fontsize=13)
pl.title("2013-2019 Overall stats: AC Milan vs UCL teams",fontsize=25,color="tomato")
```

## PERFORMING THE LINEAR REGRESSION:

We then perform predictions based on previous results and statistics. 

In this case, we look at Milan's last season's Goal difference (19), number of Wins (19 also), and ask ourselves: 

If they added the goals a potential recruit was involved in to their season, how many wins would they have if the remaining stats (Goals conceded) remained the same? 

In other words if they improved their attack and kept the same exact defensive stats, how many wins would they have?

As stated, we see that for the 2018-2019 season Milan won 19 games, scored 55 goals and conceded 36, which left them with a goal difference of 19:

We then create a function which adds the goals a potential recruit was involved in to Milan's goal difference amount (we add the goals and assists of Eintracht Frankfurt's forward Ante RebiÄ, who scored 9 goals and had 4 assists, thus was involved in 13 goals that season):

```python
def GoalDifference():
    Goal_Difference = 19
    name = input("Insert player's name: ")
    
    def playerinfluence():
        PotentialRecruits['Goals involved'] = PotentialRecruits['Goals'] + PotentialRecruits['Assists']
        return PotentialRecruits[PotentialRecruits['Full name'].str.contains(name)][['Goals involved']]
    
    PotentialRecruits['New goal difference'] = Goal_Difference + PotentialRecruits[PotentialRecruits['Full name'].str.contains(name)][['Goals involved']]
    return PotentialRecruits[PotentialRecruits['Full name'].str.contains(name)][['New goal difference']]
```

```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm


X = Data_ACMilan[['Goal difference']] # here we have 1 variable for linear regression.
Y = Data_ACMilan['Wins']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

#linear regression equation:
print('\nLinear Regression equation:\n', regr.intercept_, 
      '+', regr.coef_, '* New Goal Difference')

# prediction with sklearn
New_Goal_Difference = 32
print ('\n With a Goal difference of', New_Goal_Difference, 
       'AC Milan is predicted to win this amount of games: \n',
       regr.predict([[New_Goal_Difference]]))
```
